## Point 클래스 메소드 - 일반

- 연산자 오버로딩을 통한 산술 연산을 구현하는 것 이외에 필요한 연산은?
- 점 길이(벡터의 길이), 내적, 외적, 두 점 사이의 거리, 점의 이동, 점의 회전 등 다양함. 아래의 메소드로 구현 가능





image.png

```python
import math

class Point:
  def __init__(self, x=0, y=0):
    self.x = x
    self.y = y

  def __str__(self):
    return f"({self.x}, {self.y})"

  def __sub__(self, other):
    return Point(self.x - other.x, self.y - other.y)

  def length(self):
    return math.sqrt(self.x**2 + self.y**2)
  
  def dot(self,q):
    return self.x*q.x + self.y*q.y

  def dist(self, q):
    return (self-q).length()

  def move(self, dx=0, dy=0):
    self.x += dx
    self.y += dy


p = Point(1,2)
q = Point(2,3)

print(f"p = {p}, q = {q}")
print('lenght of p=', p.length())
print('dot of p and q =', p.dot(q))
print('dis of p and q=', p.dist(q))
p.move(3,5)
print('move p by (3,5)=', p)
```

- 점의 좌표 값을 읽거나 변경하고 싶다면, p.x, p.y 처럼 멤버 값 직접 참조 가능(그러나 객체 지향 언어 원칙에 위배됨)
- 최대한 클래스 내부의 멤버값을 직접 참조하지 않고 정해진 메소드로만 참조하는 것이 좋음

```python
import math

class Point:
  def __init__(self, x=0, y=0):
    self.x = x
    self.y = y

  def __str__(self):
    return f"({self.x}, {self.y})"

  def __sub__(self, other):
    return Point(self.x - other.x, self.y - other.y)

  def __lenght__(self):
    return math.sqrt(self.x**2 + self.y**2)

  def dot(self, q):
    return self.x*q.x + self.y*q.y

  def dist(self,q):
    return (self-q).length()

  def move(self, dx=0, dy=0):
    self.x += dx
    self.y += dy
    
  def getX(self):
    return self.x

  def getY(self):
    return self.y

  def get(self):
    return self.x, self.y

  def setX(self, val):
    self.x = val
    return None

  def setY(self, val):
    self.y = val
    return None

  def set(self, ox,oy):
    self.x = ox
    self.y = oy
    return None
```

```python

p = Point(1,2)
q = Point(2,3)

print(p)
print(q)
```

```python
print(p.getX())
print(p.getY())
print(p.get())
```
